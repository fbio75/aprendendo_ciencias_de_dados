```python
df = sns.load_dataset('tips')
```

```python
sns.countplot(data=df, x = 'sex', hue='smoker')
plt.title('Homens e mulheres fumantes e não fumantes')
plt.show()
```

```python
dados = df.groupby(['sex', 'smoker']).size().unstack()
print(dados)
```

```python
dados.plot(kind='bar')
plt.show()
```

```python
dados.plot(kind='line')
plt.show()
```

```python
dados.plot(kind='hist')
plt.show()
```

```python
dados['Yes'].plot(kind='pie')
plt.show()
```

```python
dados['Yes'].plot(
    kind='pie',
    autopct='%1.1f%%'
)
dados['Yes'].plot(
    kind='pie',
    autopct='%1.1f%%'
)
```

```python
dados.plot(kind='bar')
plt.show()
```

```python
dados.plot(kind='bar', figsize=(8,5))
plt.show()
```

```python
dados.plot(kind='bar', rot=0)
```

Qual sexo dá maiores gorjetas?

```python
df.groupby('sex')['tip'].mean().round(2)
```

```python
df.groupby('sex')['tip'].mean().round(2).plot(kind='bar')
plt.show()
```

Qual dia tem maior faturamento?

```python
df.groupby('day')['total_bill'].sum()
```

```python
df.groupby('day')['total_bill'].sum().round(2).plot(kind='bar')
plt.show()
```

Quantidade de fumantes e não fumantes

```python
df['smoker'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')
plt.show()
```

Qual horário possui mais clientes?

```python
df['time'].value_counts().plot(kind='bar')
plt.show()
```

Média de contas por horário

```python
df.groupby('time')['total_bill'].mean().plot(kind='bar')
plt.show()
```

Distribuição das contas

```python
df['total_bill'].plot(kind='hist')
plt.show()
```

Relação entre conta e gorjeta

```python
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('conta')
plt.ylabel('gorjeta')
plt.show()
```

Comparar fumantes e não fumantes

```python
df.groupby('smoker')['total_bill'].mean().plot(kind='bar')
plt.show()
```

Praticar filtros

```python
df[df['tip'] > 5]
```

```python
df[df['size'] >= 3]
```

Combinar filtros
